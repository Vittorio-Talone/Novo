```python
dados_quantitativos = pnsestilo['C008']

media = dados_quantitativos.mean()
mediana = dados_quantitativos.median()
moda = dados_quantitativos.mode()[0]
quartis = dados_quantitativos.quantile([0.25, 0.5, 0.75])
amplitude = dados_quantitativos.max() - dados_quantitativos.min()

print(f"Média: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
print(f"Quartis: {quartis}")
print(f"Amplitude: {amplitude}")
```

## Breves intepretações a partir dos dados analisados:

**Média** de 34 anos: A idade central tende a se aproximar desse valor. Deve-se considerar que a média é afetada por valores extremos (idades muito altas ou muito baixas).

**Mediana** como 33.0 anos: Metade das pessoas tem 33 anos ou menos e a outra metade tem 33 anos ou mais.

A mediana está bem próxima da média. Em geral, isto indica que não temos muitos valores extremos.

**Moda** como 31 anos: A idade de 31 anos foi a mais comum entre as 16 pessoas.

**Quartis**:

*1º Quartil* (Q1 - 29.75): Aproximadamente 25% das pessoas têm idades abaixo de 29,75 anos.

*2º Quartil* (Q2 - 33.0): 50% das pessoas têm idades abaixo de 33 anos.

*3º Quartil* (Q3 - 36.25): Aproximadamente 75% das pessoas têm idades abaixo de 36,25 anos e 25% das pessoas têm idades acima de 36,25 anos.

**Amplitude** de 39 anos: variação de idade no grupo. Este alto valor indica que temos tanto pessoas muito jovens quanto pessoas mais velhas.
